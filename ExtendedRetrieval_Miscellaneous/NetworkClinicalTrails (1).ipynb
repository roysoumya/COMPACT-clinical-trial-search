```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

# 1. Create the graph
g = nx.from_pandas_edgelist(da, source='reference_pmid', target='nct_id') 

# 2. Create a layout for our nodes 
layout = nx.spring_layout(g,iterations=50)

# 3. Draw the parts we want
# Edges thin and grey
# People small and grey
# Clubs sized according to their number of connections
# Clubs blue
# Labels for clubs ONLY
# People who are highly connected are a highlighted color

# Go through every club name, ask the graph how many
# connections it has. Multiply that by 80 to get the circle size
clt_size = [g.degree(nct_id) * 80 for nct_id in clt]
nx.draw_networkx_nodes(g, 
                       layout, 
                       nodelist=clt, 
                       node_size=clt_size, # a LIST of sizes, based on g.degree
                       node_color='lightblue')

# Draw EVERYONE
nx.draw_networkx_nodes(g, layout, nodelist=ref, node_color='#cccccc', node_size=100)

# Draw POPULAR PEOPLE
#popular_people = [reference_pmid for reference_pmid in ref if g.degree(reference_pmid) > 1]
#x.draw_networkx_nodes(g, layout, nodelist=popular_people, node_color='orange', node_size=100)

#nx.draw_networkx_edges(g, layout, width=1, edge_color="#cccccc")

node_labels = dict(zip(clt, clt))
nx.draw_networkx_labels(g, layout, labels=node_labels)

# 4. Turn off the axis because I know you don't want it
plt.axis('off')

plt.title("CLINICAL_TRAILS")

# 5. Tell matplotlib to show it
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

# 1. Create the graph
g = nx.from_pandas_edgelist(da, source='reference_pmid', target='nct_id') 

# 2. Create a layout for our nodes 
layout = nx.spring_layout(g,iterations=50)

# 3. Draw the parts we want
nx.draw_networkx_edges(g, layout, edge_color='#AAAAAA')

clt = [node for node in g.nodes() if node in da.nct_id.unique()]
size = [g.degree(node) * 80 for node in g.nodes() if node in da.nct_id.unique()]
nx.draw_networkx_nodes(g, layout, nodelist=clt, node_size=size, node_color='lightblue')

ref = [node for node in g.nodes() if node in da.reference_pmid.unique()]
nx.draw_networkx_nodes(g, layout, nodelist=ref, node_size=100, node_color='#AAAAAA')

high_degree_people = [node for node in g.nodes() if node in da.reference_pmid.unique() and g.degree(node) > 1]
nx.draw_networkx_nodes(g, layout, nodelist=high_degree_people, node_size=100, node_color='#fc8d62')

clt_dict = dict(zip(clt, clt))
nx.draw_networkx_labels(g, layout, labels=clt_dict)

# 4. Turn off the axis because I know you don't want it
plt.axis('off')

plt.title("CLINICAL_TRIALS")

# 5. Tell matplotlib to show it
plt.show()
```
