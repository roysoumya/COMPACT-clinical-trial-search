```python
from quickumls import QuickUMLS
matcher = QuickUMLS('/home/gunjan/Documents/QuickUMLS_files/')
```

```python
import os
dir_name = 'ZeroShot'
list_dec = [os.path.join(dir_name,'decoded',x) for x in os.listdir('%s/decoded/'%dir_name)]
list_ref = [os.path.join(dir_name,'reference',x) for x in os.listdir('%s/reference/'%dir_name)]

list_dec = sorted(list_dec,key = lambda x : int(x.split('/')[-1].split('_')[0]))
list_ref = sorted(list_ref,key = lambda x : int(x.split('/')[-1].split('_')[0]))
```

```python
from collections import defaultdict
def getMatchInfo(list_matches):
    dict_concepts = defaultdict(list)
    for match_l in list_matches:
        for match_l_d in match_l:
            n_gram = match_l_d['ngram']
            n_gram = n_gram.strip() 
                  
            if '\n' in n_gram: continue 

            for stype in match_l_d['semtypes']: 
                if not stype in dict_concepts[n_gram]: dict_concepts[n_gram].append(stype) 

    
    str_ret = str()
    for key,val in dict_concepts.items():
        str_ret += key + ':'
        for c_name in val:
            if not '\n' in c_name:
                str_ret += c_name+'|'
        str_ret += '\n'
    return (str_ret)
      
```

```python
list_id_dec,list_dec_sum,list_ref_sum,list_tags_dec, list_tags_ref =list(), list(),list(), list(), list()

for fname1,fname2 in zip(list_dec,list_ref):
    
    print('++++++++++++++++++',fname1,fname2)
    
    dec_id = fname1.split('/')[-1].split('_')[0]
    ref_id = fname2.split('/')[-1].split('_')[0]
    
    assert dec_id == ref_id
    
    with open(fname1,'r') as f:
        str_abs = f.read().lower()
    f.close()

    with open(fname2,'r') as f:
        str_ref = f.read().lower()
    f.close()
    
    list_match_dec = matcher.match(str_abs, best_match=True, ignore_syntax=False)
    list_match_ref = matcher.match(str_ref, best_match=True, ignore_syntax=False)
    
    str_match_dec = getMatchInfo(list_match_dec)
    str_match_ref = getMatchInfo(list_match_ref)
    
    list_tags_dec.append(str_match_dec)
    list_tags_ref.append(str_match_ref) 
    list_dec_sum.append(str_abs)
    list_ref_sum.append(str_ref)
    list_id_dec.append(dec_id)
```

```python
d_dump = {'ids' : list_id_dec,
              'Dec_Summ': list_dec_sum,
              'Ref_Summ': list_ref_sum,
              'Stype_Info_dec': list_tags_dec,
              'Stype_Info_ref': list_tags_ref
             }
```

```python
import pandas as pd
df_dump = pd.DataFrame(d_dump)
df_dump = df_dump.dropna()
df_dump.to_csv('%s/Stype_Info_Novel.csv'%dir_name,index=False)
```
