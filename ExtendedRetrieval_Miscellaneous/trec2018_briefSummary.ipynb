```python
from collections import defaultdict
query_trials_dict = defaultdict(list)
import re
```

```python
for file in onlyfiles:
    x = re.findall('[0-9]+', file)
    filename = x[0]
    csv_data = pd.read_csv(join(mypath, file))
    query_trials_dict[filename] = csv_data['nct_id'].tolist()
```

```python
qrel_trec_set = pd.read_csv('COCTR_multidimensional_ranking-master/goldStandardDataTREC_2018/data/qrel_relevant_trec2018_all.csv')
```

```python
print(qrel_trec_set.columns)
print(qrel_trec_set.shape)
qrel_trec_set.head()
```

```python
retr_trials_dict_1 = defaultdict(list)
retr_trials_dict_2 = defaultdict(list)
retr_trials_dict_0 = defaultdict(list)
```

```python
for row_id in range(qrel_trec_set.shape[0]):
    key = str(qrel_trec_set.iloc[row_id, 0])
    if qrel_trec_set.iloc[row_id, 2] == 1:
        retr_trials_dict_1[key].append(qrel_trec_set.iloc[row_id, 1])
    elif qrel_trec_set.iloc[row_id, 2] == 0:
        retr_trials_dict_0[key].append(qrel_trec_set.iloc[row_id, 1])
    else:
        retr_trials_dict_2[key].append(qrel_trec_set.iloc[row_id, 1])
```

```python
retr_trials_dict_0
```

```python
key_list = list()
num_qrel_trials = list()
num_retr_trials_0 = list()
num_retr_trials_1 = list()
num_retr_trials_2 = list()
num_itrsct_trials_0 = list()
num_itrsct_trials_1 = list()
num_itrsct_trials_2 = list()
```

```python
for key_id in range(50):
    key_id = str(key_id)
    key_list.append(key_id)
    qrel_trials = set(query_trials_dict[key_id])
    retr_trials_0 = set(retr_trials_dict_0[key_id])
    retr_trials_1 = set(retr_trials_dict_1[key_id])
    retr_trials_2 = set(retr_trials_dict_2[key_id])
    
    #print(qrel_trials)
    inter_trials_0 = qrel_trials.intersection(retr_trials_0)
    inter_trials_1 = qrel_trials.intersection(retr_trials_1)
    inter_trials_2 = qrel_trials.intersection(retr_trials_2)
    
    num_qrel_trials.append(len(qrel_trials))
    num_retr_trials_0.append(len(retr_trials_0))
    num_retr_trials_1.append(len(retr_trials_1))
    num_retr_trials_2.append(len(retr_trials_2))
    
    num_itrsct_trials_0.append(len(inter_trials_0))
    num_itrsct_trials_1.append(len(inter_trials_1))
    num_itrsct_trials_2.append(len(inter_trials_2))
```

```python
coverage_trec_2018 = pd.DataFrame({'query_id': key_list, 'num_qrel_coctr': num_qrel_trials, 'num_gold_rele0': num_retr_trials_0, 'num_gold_rele1': num_retr_trials_1, 'num_gold_rele2': num_retr_trials_2, 'num_inter_0': num_itrsct_trials_0, 'num_inter_1': num_itrsct_trials_1, 'num_inter_2': num_itrsct_trials_2})
coverage_trec_2018
```
